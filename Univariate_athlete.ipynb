```python
medalist_df= data_df.loc[~data_df['Medal'].isnull()]
medalist_df
```

```python
def plot_col(df,col,chart_type,dtype=int):
    col_df=df[col]
    col_df=col_df.dropna()
    col_df=col_df.astype(dtype)
    if chart_type=="Histogram":
        sns.countplot(col_df)
    if chart_type=="Density":
        sns.distplot(col_df)
```

```python
plot_col(data_df,'Age',"Histogram")
```

```python
from scipy.stats import skew
Age_df=medalist_df['Age']
Age_df=Age_df.dropna()
Age_df=Age_df.astype(int)
print(skew(Age_df))
```
